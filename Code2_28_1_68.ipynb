```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('insurance.csv')
df.head()
```

```python
df.columns
```

```python
# กำจัดค่าว่าง
df.dropna(inplace = True)
```

```python
input_columns = ['age', 'bmi', 'children'] #แก้
output_columns = ['charges'] #แก้
```

```python
X = df.loc[:, input_columns]
y = df.loc[:, output_columns]
```

```python
# splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .20, random_state = 1,)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
model1 = LinearRegression() #สร้าง model : y=mx + b
model1.fit(X_train, y_train) #เรียนหาค่า m กับ b ออกมา
```

```python
#แสดงค่า m และ b
print(model1.intercept_) #b จุดตัดแกน y
print(model1.coef_) #m ความชัน
```

```python
#ทำนายบนชุดทดสอบ
y_p = model1.predict(X_test)
#วัดผล MSE
from sklearn import metrics
print(np.sqrt(metrics.mean_squared_error(y_test, y_p)))
```

```python
#ใช้ model ที่ได้ ทำนายเบี้ยประกัน #แก้
#age = 30, bmi=35, children =1
#age = 19, bmi=25, children =0
#age = 41, bmi=23, children =2
X_quiz = [[30,35,1],[19,25,0], [41, 23, 2]]
y_quiz = model1.predict(X_quiz)
print(y_quiz)
```

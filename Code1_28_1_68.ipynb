```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#อ่านข้อมูลจากไฟล์ study.csv
df = pd.read_csv('study.csv')
df.head()
```

```python
#ตรวจสอบว่ามีคอลัมอะไรบ้าง
df.columns
```

```python
input_columns = ['Hours'] #แก้
output_columns = ['Scores'] #แก้
```

```python
x = df.loc[:,input_columns]
y = df.loc[:,output_columns]
```

```python
#X.shape
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = .20, random_state = 1,)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
model1 = LinearRegression() #สร้าง model : y=mx + b
model1.fit(X_train, y_train) #เรียน หาค่า m กับ b ออกมา
```

```python
#แสดงค่า m และ b
print(model1.intercept_) #b จุดตัดแกน y
print(model1.coef_) #m ความชั้น
```

```python
#ทำนายบนชุดทดสอบ
y_p = model1.predict(X_test)
```

```python
#วัดผล MSE
from sklearn import metrics
print(np.sqrt(metrics.mean_squared_error(y_test, y_p)))
```

```python
#หากเรียน 7.5, 8.5, 9.25
X_quiz = [7.5], [8.5], [9.25]
y_quiz = model1.predict(X_quiz)
print(y_quiz)
```

```python
#ใช้ model ที่ได้ หากเรียน 12.2 , 15.3, 1, 0,6,7.2
#จะได้คะแนนเท่าใด
X_quiz = [[12.2], [15.3], [1] , [0] , [6], [7.2]]
y_quiz = model1.predict(X_quiz)
print(y_quiz)
```
